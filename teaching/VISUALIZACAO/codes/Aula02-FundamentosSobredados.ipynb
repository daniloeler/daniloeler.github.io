```python
data2 = pd.DataFrame([['P1',1500.89,30,87.6,1,0],
                     ['P2',789.52,48,None,2,0],
                     ['P3',1000.00,28,70.5,2,1],
                     ['P4',589.36,39,90.5,3,1],
                     [np.nan,None,None,None,None,None],
                     ['P6',589.36,39,90.5,None,None]])
data2
```

- Substituindo dados ausentes

```python
data4 = pd.DataFrame([['P1',1500.89,30,87.6,1,0],
                     ['P2',789.52,48,np.nan,2,0],
                     ['P3',1000.00,28,70.5,2,1],
                     ['P4',589.36,39,90.5,3,1],
                     ['P5',900.00,np.nan,55.6,np.nan,np.nan],
                     ['P6',589.36,39,90.5,np.nan,np.nan]])
data4
```

```python
data_zero = data4.fillna(0) #substitui NAN por um valor
data_zero
```

```python
data4.describe() #medidas estatísticas do dataframe -- desconsidera o NAN
```

```python
#usar dicionario para indicar valores distintos para cada atributo
#substitui NAN pelo valor da media -- conforme apresentado na célula anterior
data_media = data4.fillna({1:894.85, 2:36.80, 3:78.94, 4: 2, 5: 1}) #tomar cuidado no booleano
data_media
```

```python
data_media.describe() #medidas estatísticas após substituição
```

```python
data4
```

```python
#modo automatico -- entretanto, deve-se tomar cuidado com o booleano -- note que ele não foi alterado nesse momento
# e deve ser alterado manualmente -- ou por algum metodo automatico de contagem ou outra estratégia
data4.fillna(data4.mean()[0:4]) #data4.mean calcula as médias por atributo e fillna preenche os dados ausentes
```

```python
data4 #note que o data4 ainda não foi modificado, pois fillna retorna um novo dataframe
```

```python
data4.fillna({1:894.85, 2:36.80, 3:78.94, 4: 2, 5: 1}, inplace=True) #altera o data4 (inplace)
data4
```

```python
data5 = pd.DataFrame([['P1',1500.89,30,87.6,1,0],
                     ['P2',789.52,48,np.nan,2,0],
                     ['P3',1000.00,28,70.5,2,1],
                     ['P4',589.36,39,90.5,3,1],
                     ['P5',900.00,np.nan,55.6,np.nan,np.nan],
                     ['P6',589.36,39,90.5,np.nan,np.nan]])
data5
```
